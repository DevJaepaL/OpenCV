```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
# 이미지 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/Lena-Gaussian-noise1.jpg')
img_2 = cv2.imread('/content/gdrive/My Drive/Image_Processing/lena.jpg')

# 컬러 이미지 -> 그레이 이미지 변환
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2GRAY)

# 가우시안 블러 적용.
blur = cv2.GaussianBlur(gray_img,(3,3),1)
blur_2 = cv2.GaussianBlur(gray_img2,(3,3),1)

# 라플라시안 에지 컨벌루션 연산
laplacian = cv2.Laplacian(blur,-1,1)
laplacian_img = laplacian/laplacian.max() # 최대값으로 정규화 필요
laplacian1 = cv2.Laplacian(gray_img,-1,1)
gau_laplacian_img = laplacian1/laplacian1.max()

laplacian_2 = cv2.Laplacian(blur_2,-1,1)
laplacian_img2 = laplacian_2/laplacian_2.max()
gau_laplacian_img = laplacian1/laplacian1.max()
gau_laplacian_img2 = laplacian_2/laplacian_2.max()

#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(2,3,1)
plt.title("Noise Image")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(2,3,2)
plt.title("Laplacian Edge Detection")
plt.imshow(laplacian_img, cmap='gray')
plt.axis("off")
plt.subplot(2,3,3)
plt.title("Gaussian-Laplacian Edge Detection")
plt.imshow(gau_laplacian_img, cmap='gray')
plt.axis("off")
plt.subplot(2,3,4)
plt.title("Original Image")
plt.imshow(gray_img2, cmap='gray')
plt.axis("off")
plt.subplot(2,3,5)
plt.title("Laplacian Edge Detection 2")
plt.imshow(laplacian_img2, cmap='gray')
plt.axis("off")
plt.subplot(2,3,6)
plt.title("Gaussian-Laplacian Edge Detection 2")
plt.imshow(gau_laplacian_img2, cmap='gray')
plt.axis("off")
plt.show()
```
