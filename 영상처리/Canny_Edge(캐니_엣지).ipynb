1차 미분 연산자인 Rober Cross, Sobel, Prewitt Edge의 단점인 잡음에 민감하다는 특성을 극복하기 위해 만든 검출기.

알고리즘의 과정

1.   가우시안 커널을 이용하여 영상의 잡음 제거
2.   에지 검출기를 이용하여 에지 검출
3.   2개의 임계값을 이용한 에지 트래킹을 이용해 강한 에지만을 추출함.







```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 

# 영상 호출
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/lena.jpg')

# 컬러 이미지 -> 그레이로 변환
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# OpenCV에서 Canny() 함수 제공함.
output_img = cv2.Canny(gray_img,150,100)

# 이미지 출력
plt.figure(figsize=(15,15))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Canny Edge Detector")
plt.imshow(output_img, cmap='gray')
plt.axis("off")
plt.show()
```
