```
# 가우시안 분포(Gaussian distribution)를 따르는 가우시안 커널을 사용하여 영상의 자븜을 제거하는 기법
# 평균 필터(uniformly distributed)와는 달리 중심의 가까운 픽셀에 더 높은 가중치를 부여한다.
# 비슷한 크기의 평균 필터에 비해 '에지 정보'를 잘 유지하면서 자연스럽게 스무딩을 적용할 수 있다.
# 대부분의 컨볼루션 기반의 스무딩 필터와 달리 세밀함이 잘 보존된다.

from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/Lena-Gaussian-noise1.jpg')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 5x5 가우시안 커널 적용하기
output_img1 = cv2.medianBlur(gray_img,5)
output_img2 = cv2.GaussianBlur(gray_img,(5,5),0)
output_img3 = cv2.GaussianBlur(gray_img,(7,7),0)

#그림을 화면에 출력
plt.figure(figsize=(30,30))# 영상의 크기를 키워주자
plt.subplot(1,4,1)
plt.title("Original Image")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,4,2)
plt.title("Median filtering Image(5x5)")
plt.imshow(output_img1, cmap='gray')
plt.axis("off")
plt.subplot(1,4,3)
plt.title("Gaussian filtering Image(5x5)")
plt.imshow(output_img2, cmap='gray')
plt.axis("off")
plt.subplot(1,4,4)
plt.title("Gaussian filtering Image(7x7)")
plt.imshow(output_img3, cmap='gray')
plt.axis("off")
plt.show()
```
