영상 불러오기

```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/Lena-Gaussian-noise1.jpg')
img2 = cv2.imread('/content/gdrive/My Drive/Image_Processing/fce(salt_pepper noise).bmp')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

중간값 커널 적용하기(mediaBlur)

```
# 블러 적용(Meaning Filter)
output_img1 = cv2.blur(gray_img,(5,5))
output_img3 = cv2.blur(gray_img2,(5,5))
# 5x5 중간값 커널 적용하기
output_img2 = cv2.medianBlur(gray_img,5)
output_img4 = cv2.medianBlur(gray_img2,5)
```

화면 출력

```
# 중간값 필터링(Median Filtering)은 주변의 픽셀 값 중 하나를 이용하여 화질 저하가 거의 없이 잡음 픽셀 제거
# 중간값 필터는 에지와 경계선과 같은 영상의 고주파 성분을 잘 보존하면서 잡음제거의 장점
# 주변 픽셀값들 중에서 중간값을 찾기 위한 추가 연산(퀵정렬) 알고리즘이 필요하여 연산과 시간이 필요함

#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(2,3,1)
plt.title("Image1")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(2,3,2)
plt.title("Mean filtering Image")
plt.imshow(output_img1, cmap='gray')
plt.axis("off")
plt.subplot(2,3,3)
plt.title("Median filtering Image")
plt.imshow(output_img2, cmap='gray')
plt.axis("off")
plt.subplot(2,3,4)
plt.title("Image2")
plt.imshow(gray_img2, cmap='gray')
plt.axis("off")
plt.subplot(2,3,5)
plt.title("Mean filtering Image 2")
plt.imshow(output_img3, cmap='gray')
plt.axis("off")
plt.subplot(2,3,6)
plt.title("Median filtering Image 2")
plt.imshow(output_img4, cmap='gray')
plt.axis("off")
plt.show()
```
