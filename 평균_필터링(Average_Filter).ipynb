1. 파일 불러오기

```
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/gdrive') 
#영상 불러오기
img = cv2.imread('/content/gdrive/My Drive/Image_Processing/Lena-Gaussian-noise1.jpg')

# color영상을 gray영상으로 만들기
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur_img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

2. 커널 만들기

```
# 계수가 1로 구성된 3x3커널 만들기
Kernel_1 = np.ones((3,3),np.float32)/9
Kernel_2 = np.ones((5,5),np.float32)/25

print(Kernel_1)
print(Kernel_2)
```

3. 필터 적용하기

```
# 만들어진 커널을 이용하여 필터 적용
output_img1 = cv2.filter2D(gray_img,-1,Kernel_1)
output_img2 = cv2.filter2D(gray_img2,-1,Kernel_2)

# 블러 이미지(커널 계수 지정 X)
output_img3 = cv2.blur(blur_img,(6,6))
output_img4 = cv2.blur(blur_img,(9,9))
```

그림 화면 출력

```
#그림을 화면에 출력
plt.figure(figsize=(20,20))# 영상의 크기를 키워주자
plt.subplot(2,3,1)
plt.title("Original IMG")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,3,2)
plt.title("3*3 Avg Filter")
plt.imshow(output_img1, cmap='gray')
plt.axis("off")
plt.subplot(1,3,3)
plt.title("5*5 Avg Filter")
plt.imshow(output_img2, cmap='gray')
plt.axis("off")
plt.show()
```

```
# 블러 이미지 출력
plt.figure(figsize=(20,20))
plt.subplot(1,3,1)
plt.title("Original IMG")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")
plt.subplot(1,3,2)
plt.title("6*6 Avg Filter(blur)")
plt.imshow(output_img3, cmap='gray')
plt.axis("off")
plt.subplot(1,3,3)
plt.title("9*9 Avg Filter(blur)")
plt.imshow(output_img4, cmap='gray')
plt.axis("off")
plt.show()
```
